```python
sorted_df=df.sort_values('popularity',ascending=True).head(10)
sorted_df
```

```python
df.describe().transpose()
```

```python
#top 10 song
most_popular=df.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
df.set_index("release_date",inplace=True)
df.index=pd.to_datetime(df.index)
df.head()
```

```python
#filter specific rows of the dataset
df[["artists"]].iloc[1229]#artist at 18th row by index location
```

```python
#converting duartion into sec from ms
df["duration"]=df["duration_ms"].apply(lambda x: round(x/1000))
df.drop("duration_ms",inplace=True,axis=1)
```

```python
df.duration.head()
```

```python
#correlation map
corr_df=df.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidth=1,linecolor="Black")
heatmap.set_title("correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df.sample(int(0.004*len(df)))
```

```python
print(len(sample_df))
```

```python
#regression b/w loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c" ).set(title="Loudness vs Energy Coorelation")
```

```python
#regression b/w loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="b" ).set(title="popularity vs acousticness Coorelation")
```

```python
df['dates']=df.index.get_level_values('release_date')
df.dates=pd.to_datetime(df.dates)
years=df.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
total_dr=df.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="year vs Duaration")
plt.xticks(rotation=90)
```
